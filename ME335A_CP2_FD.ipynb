```python
# Give path to save the mesh file
meshPath = '/content/drive/MyDrive/'

def CP2mesh(Hmax, elementOrder, elementType):
  # Given Hmax, construct a mesh to be read by Firedrake
  gmsh.initialize()
  gmsh.model.add('cp2Mesh')
  meshObject = gmsh.model

  # Add points
  # Points for the outer boundary
  point1 = meshObject.geo.addPoint(-1,-1,0,Hmax, 1)
  point2 = meshObject.geo.addPoint(1,-1,0,Hmax, 2)
  point3 = meshObject.geo.addPoint(1,1,0,Hmax, 3)
  point4 = meshObject.geo.addPoint(-1,1,0,Hmax, 4)

  # Points for the central hole
  point5 = meshObject.geo.addPoint(0,-0.25,0,Hmax, 5)
  point6 = meshObject.geo.addPoint(0.25,0,0,Hmax, 6)
  point7 = meshObject.geo.addPoint(0,0.25,0,Hmax, 7)
  point8 = meshObject.geo.addPoint(-0.25,0,0,Hmax, 8)

  # Points for the other holes
  point9 = meshObject.geo.addPoint(-0.5,0.4,0,Hmax, 9)
  point10 = meshObject.geo.addPoint(-0.4,0.5,0,Hmax, 10)
  point11 = meshObject.geo.addPoint(-0.5,0.6,0,Hmax, 11)
  point12 = meshObject.geo.addPoint(-0.6,0.5,0,Hmax, 12)

  point13 = meshObject.geo.addPoint(0.5,0.4,0,Hmax, 13)
  point14 = meshObject.geo.addPoint(0.6,0.5,0,Hmax, 14)
  point15 = meshObject.geo.addPoint(0.5,0.6,0,Hmax, 15)
  point16 = meshObject.geo.addPoint(0.4,0.5,0,Hmax, 16)

  point17 = meshObject.geo.addPoint(0.5,-0.6,0,Hmax, 17)
  point18 = meshObject.geo.addPoint(0.6,-0.5,0,Hmax, 18)
  point19 = meshObject.geo.addPoint(0.5,-0.4,0,Hmax, 19)
  point20 = meshObject.geo.addPoint(0.4,-0.5,0,Hmax, 20)

  point21 = meshObject.geo.addPoint(-0.5,-0.6,0,Hmax, 21)
  point22 = meshObject.geo.addPoint(-0.4,-0.5,0,Hmax, 22)
  point23 = meshObject.geo.addPoint(-0.5,-0.4,0,Hmax, 23)
  point24 = meshObject.geo.addPoint(-0.6,-0.5,0,Hmax, 24)

  # Construct lines from points
  line1 = meshObject.geo.addLine(1, 2, 101)
  line2 = meshObject.geo.addLine(2, 3, 102)
  line3 = meshObject.geo.addLine(3, 4, 103)
  line4 = meshObject.geo.addLine(4, 1, 104)

  line5 = meshObject.geo.addLine(5, 6, 105)
  line6 = meshObject.geo.addLine(6, 7, 106)
  line7 = meshObject.geo.addLine(7, 8, 107)
  line8 = meshObject.geo.addLine(8, 5, 108)

  line9 = meshObject.geo.addLine(9, 10, 109)
  line10 = meshObject.geo.addLine(10, 11, 110)
  line11 = meshObject.geo.addLine(11, 12, 111)
  line12 = meshObject.geo.addLine(12, 9, 112)

  line13 = meshObject.geo.addLine(13, 14, 113)
  line14 = meshObject.geo.addLine(14, 15, 114)
  line15 = meshObject.geo.addLine(15, 16, 115)
  line16 = meshObject.geo.addLine(16, 13, 116)

  line17 = meshObject.geo.addLine(17, 18, 117)
  line18 = meshObject.geo.addLine(18, 19, 118)
  line19 = meshObject.geo.addLine(19, 20, 119)
  line20 = meshObject.geo.addLine(20, 17, 120)

  line21 = meshObject.geo.addLine(21, 22, 121)
  line22 = meshObject.geo.addLine(22, 23, 122)
  line23 = meshObject.geo.addLine(23, 24, 123)
  line24 = meshObject.geo.addLine(24, 21, 124)

  # Construct closed curve loops
  outerBoundary = meshObject.geo.addCurveLoop([line1, line2, line3, line4], 201)
  hole1 = meshObject.geo.addCurveLoop([line5, line6, line7, line8], 202)
  hole2 = meshObject.geo.addCurveLoop([line9, line10, line11, line12], 203)
  hole3 = meshObject.geo.addCurveLoop([line13, line14, line15, line16], 204)
  hole4 = meshObject.geo.addCurveLoop([line17, line18, line19, line20], 205)
  hole5 = meshObject.geo.addCurveLoop([line21, line22, line23, line24], 206)

  # Define the domain as a 2D plane surface with holes
  domain2D = meshObject.geo.addPlaneSurface([outerBoundary,hole1,hole2,hole3,hole4,hole5], 301)

  # Synchronize gmsh
  meshObject.geo.synchronize()

  # Add physical groups for firedrake
  meshObject.addPhysicalGroup(2, [301], name='domain')

  meshObject.addPhysicalGroup(1, [line2], 1)
  meshObject.addPhysicalGroup(1, [line3], 2)
  meshObject.addPhysicalGroup(1, [line4], 3)
  meshObject.addPhysicalGroup(1, [line1], 4)
  meshObject.addPhysicalGroup(1, [line5, line6, line7, line8], 5)
  meshObject.addPhysicalGroup(1, [line9, line10, line11, line12], 6)
  meshObject.addPhysicalGroup(1, [line13, line14, line15, line16], 7)
  meshObject.addPhysicalGroup(1, [line17, line18, line19, line20], 8)
  meshObject.addPhysicalGroup(1, [line21, line22, line23, line24], 9)

  # Set element order
  meshObject.mesh.setOrder(elementOrder)

  if elementType == 2:
    # Generate quad mesh from triangles by recombination
    meshObject.mesh.setRecombine(2, domain2D)

  # Generate the mesh
  gmsh.model.mesh.generate(2)

  gmsh.write(meshPath + 'cp2FiredrakeMeshFile.msh')

  gmsh.finalize()

  return
```

```python
# Generate the mesh
elementOrder = 1 # Polynomial order in each element (integer)
elementType = 2 # 1 - Triangle; 2 - Quad
HMax = 0.8

CP2mesh(HMax, elementOrder, elementType)
```
